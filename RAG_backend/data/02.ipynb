- Explicitly program rules
    - Rules-based approach
    - Probability of default in banking

## Machine Learning Use Cases Over Traditional Approach

1. As a substitute for learning and fine-tuning a long list of rules.  ML can simplify rules, remove unimportant variables (feature engineering and selection) and do a better job of generalizing than a rules-based approach.
2. As a solution to complex problems with many variables.  Too difficult to continue to write long list of rules to incorporate all of the complexity of speech and text, for example.
3. As a solution to a continually fluctuating environments.  Too difficult to continually rewrite rules as data streams in by the TB (or higher) every minute
4. As a way to gain new insights from the data about the relationships between variables within large amounts of data

With a machine learning system you **don't** have to:

1. Write a long list of rules.  It will learn them for you.  An ML system is automatic.
2. Try to write a long list of very **complicated** rules and ensure they work together and do not contradict one another.  An ML system deals well with complexity while trying to find simplicity




3. Continually update the system.  An ML system can be easily, continually and automatically updated



4. Draw insights from the data manually.  An ML system can be inspected to understand the relationships it has learned and allow the user to learn from those relationships.  [Statistical learning](https://www.statlearning.com/) vs. Machine Learning




## Types of Machine Learning

### Training Supervision

#### Supervised Learning

- Supervised learning is where our data is labeled or has a desired solution.  We use features to predict that label.

- Two, most common, types of supervised learning
    1. Classification - outcomes (a.k.a. classes or labels) are categories
    2. Regression - outcomes (a.k.a. targets) are numeric

- Sometimes the word label is used synonymously with target, but more commonly label is used for classification and target for regression.

#### Unsupervised Learning

- Unsupervised learning is where our data has no label or desired solution.  We use different techniques to explore relationships in our data without creating a model to predict things.

Some common types of unsupervised learning problems

1. Clustering



2. Visualization



3. Dimensionality reduction - reduce or combine similar features (e.g. age of car and number of miles)
4. Anomaly detection - outlier identification
5. Novelty detection - identify new data not seen during training - need a clean data set for training
6. Association rule learning - *if you bought this, you might also like to buy this*

#### Semi-supervised Learning

You have a little labeled data and a lot of unlabelled data.  You can used unsupervised learning to label the unlabelled data and supervised to train once you have labelled all the data

1. Perform clustering on the unlabelled data
2. Find the number of clusters and label the data point closest to each cluster center 
3. Propagate that label to all other data points in that cluster (you can also remove any data points that are outliers at the same time)
4. Use this labelled data and the actual labelled data to train a machine learning model

#### Self-supervised Learning

Create a labelled dataset from a fully unlabelled one

- Create "holes" in an image by masking and predict the masked image portion
- Remove words from sentences and predict the missing word or just predict the next word in a sentence
- Can also be useful in training a model that can be used for other tasks.  Can fine-tune the model on other tasks.  Powerful neural network models can be trained to fill the masked words and then fine-tuned on other text classification tasks.  This is called ***transfer learning***

#### Reinforcement Learning

- Train an agent to respond to rewards and penalties
- It must learn what the best strategy is (a.k.a. policy)



## Batch Vs. Online Learning

### Batch Learning

- A.K.A Offline Learning (bad name)
- Train your model with all data at once and deploy.  Need to train new versions of the model from scratch which can be resource intensive.
- This is why it is so important to automate as much as possible
- Simple, if you model is small.  Complex, if you have a large model that should be updated frequently
- Will experience **model rot** or **model drift**, if you don't update the model and keep it fresh

### Online Learning

- A.K.A Incremental Learning
- Train a model initially and continue to train only incrementally when new training data become available (train it in mini-batches)
- Good, if you have limited computing resources
- Need to monitor this automation.  Your model could train on bad data without you knowing it.

#### Out-Of-Core Learning

- Online learning algorithms can be used to train models on enormous data sets.  Only read a few observations in at a time to not have memory overflow.
- Large language models will do just this (tensorflow, pytorch)

#### Learning Rate

- Rate at which your models learns from new observations

1. High Learning Rate:  will learn new patterns quickly, but may forget old patterns
2. Low Learning Rate: will learn new patterns more slowly and may do a better job of remembering old patterns in the data

## Instance-Based Learning vs. Model-Based Learning

Two types of machine learning model based on how predictions are made

### Instance-Based

- You make predictions by comparing a new, unlabelled instance to all the other data and label it according to the most similar instances. The system learns the examples by heart, then generalizes to new cases by using a similarity measure to compare them to the learned examples (or a subset of them)



### Model-Based

- Create a function that takes in data and produces an output.  Throw the training data away once done.



- Need to go through a process of ***model selection***
- Need to fit ***parameters*** for your model/function (variables that control how a model behaves and what results it produces).  They are internal to the model
- Need to either define a ***utility function*** (how good your model is) or a **cost function** (how bad your model is)
- Need to find a way to maximize your utility function or minimize your cost function.
- Need to go through a process of choosing parameters that maximize your utility function or minimize your cost function.  This is ***training*** your model
- Finally, can make predictions on new cases (a.k.a. inference)

### Four Parts of Supervised Machine Learning Algorithm

1. A feature representation of the input or ***data***.
2. A model that calculates an output from the inputs
3. An objective function for learning (utility/likelihood/fitness or cost function) that compares the model calculated output to the actual output
4. An algorithm for optimizing the objective function

### Main Challenges of Machine Learning

#### Insufficient Quantity of Training Data

- Need lots of data, especially for image recognition and natural language processing
- In fact, once you have enough data, many different models will perform the same



#### Nonrepresentative Training Data

- Sample too small - sample noise
- Bad sample - sample bias

#### Poor-Quality Data

- You have lots of outliers, errors and noise
- You have missing data (may need imputation)

#### Irrelevant Features

- You have lots of columns that are irrelevant.  Some approaches to machine learning can result in a lot of features
    - NLP: your features are the word counts in your document.  You may only have 100 documents, but a vocabulary of 10,000
    - Image Recognition: a single 200X200 pixel image has 40000 features and that is only if it is in black and white.

Solved by:

- Feature selection - selecting only the best features
- Feature engineering - creating new features from existing ones
- Create new features by gathering new data

#### Overfitting the Training Data

- Your model learns the training data too well and fails to ***generalize*** to new data
- More powerful model with lots of features can overfit the data

Solutions:

- Simplify the model by selecting one with fewer parameters (e.g., a linear model rather than a high-degree polynomial model), by reducing the number of attributes in the training data, or by constraining the model.
- Gather more training data.
- Reduce the noise in the training data (e.g., fix data errors and remove outliers).
- Constrain the model using **regularization**.
    - Maybe the weights of our model are only allowed to grow so big (l1 and l2 regularization)
    - Essentially make the model dumber to ensure that it works better on unseen data
    - Can configure regularization through the use of **hyperparameters**; parameters outside the model that have to be adjusted to find the best model.

#### Underfitting the Training Data

- Can also underfit or develop a model that is too simple to understand the data

Solutions:

- Select a more powerful model, with more parameters.
- Feed better features to the learning algorithm (feature engineering).
- Reduce the constraints on the model (for example by reducing the regularization hyperparameter).

#### Bias-Variance Trade Off

- Often you will struggle with the bias-variance trade-off
- From [wikipedia](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff): "Bias is an error from erroneous assumptions in the learning algorithm. High bias can cause an algorithm to miss the relevant relations between features and target outputs (underfitting)."  A simple model like simple linear regression is highly biased because you are making lots of assumptions about the underlying data.
- From [wikipedia](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff): "The variance is an error from sensitivity to small fluctuations in the training set. High variance may result from an algorithm modeling the random noise in the training data (overfitting).".  A highly complex model with lots of parameters (neural networks) have high variance and are likely to overfit the data
- Your model might be highly biased and underfit or have high variance and overfit

#### Prediction Accuracy vs. Prediction Variance

- Have to balance as well the tradeoff between prediction accuracy and prediction variance.
- Suppose you have 2 models.  One has an average accuracy of 90% and a accuracy standard deviation of 10% and another has an accuracy of 89% and a standard deviation of 2%.  Which would you prefer?

#### Data Mismatch

- You have training data that is different from the data you will perform inference on

### Testing and Validating

- Need to know how your model will generalize
- Create a training and test data split
- Train the model on the training data and measure the test/generalization/out-of-sample error on the test data

#### Hyperparameter Tuning and Model Selection

- Many times you need to fit hyperparameters
- Can do this with a holdout set.
    - Split the data 3 ways; training, dev and test
    - Train a model on the training
    - Test and the model on the dev and select the best hyperparameter
    - Choose the best model and train a final model on both the training and dev sets
    - Finally, test this model on the test data
- Best solution is to use cross validation; which we will cover next class

### $\operatorname{RMSE}(\mathbf{X}, h)=\sqrt{\left.\frac{1}{m} \sum_{i=1}^m (h\left(\mathbf{x}^{(i)}\right)-y^{(i)}\right)^2}$

# Norms

- The root of a sum of squares is the $l_2$ norm (RMSE)
- The sum of absolute values is the $l_1$ norm (MAE)
- Generally speaking, the $l_k$ norm is
    - ## $||v||_k = (|v_1|^k+|v_2|^k+...+|v_n|^k)^{1/k}$

```python
from sklearn.compose import ColumnTransformer

num_attribs = ['latitude', 'longitude', 'minimum_nights', 'number_of_reviews',
               'reviews_per_month', 'calculated_host_listings_count',
               'availability_365', 'number_of_reviews_ltm']
cat_attribs = ["neighbourhood_group"]

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
])
```

```python
X = df.drop('price', axis=1).copy()
y = df['price']
X = X.drop(['name', 'host_name', 'id', 'host_id', 'license', 'last_review'], axis=1)
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2, stratify=X['neighbourhood_group'], random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = make_pipeline(preprocessing, LinearRegression())
lin_reg.fit(X_train, y_train)
```

```python
pred = lin_reg.predict(X_train)
pred[:5].round(2)
```

```python
from sklearn.metrics import root_mean_squared_error
lin_rmse = root_mean_squared_error(y_train, pred)
lin_rmse
```

```python
pred_test = lin_reg.predict(X_test)
lin_rmse_test = root_mean_squared_error(y_test, pred_test)
lin_rmse_test
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = make_pipeline(preprocessing, DecisionTreeRegressor(random_state=42))
tree_reg.fit(X_train, y_train)
```

```python
pred = tree_reg.predict(X_train)
tree_rmse = root_mean_squared_error(y_train, pred)
tree_rmse
```

```python
pred_test = tree_reg.predict(X_test)
tree_rmse_test = root_mean_squared_error(y_test, pred_test)
tree_rmse_test

#this ends up with overfitting; bcoz of diff between res in test & train
```

```python
from sklearn.model_selection import cross_val_score

tree_rmses = -cross_val_score(tree_reg, X_train, y_train,
                              scoring="neg_root_mean_squared_error", cv=10)
```

```python
pd.Series(tree_rmses).describe()
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = make_pipeline(preprocessing,
                           RandomForestRegressor(random_state=42))
forest_rmses = -cross_val_score(forest_reg, X_train, y_train,
                                scoring="neg_root_mean_squared_error", cv=10)
```

# K Fold Cross Validation



```python
from sklearn.model_selection import GridSearchCV

full_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("random_forest", RandomForestRegressor(random_state=42)),
])
param_grid = [
    {'random_forest__max_features': [4, 6, 8, 10]}
]
grid_search = GridSearchCV(full_pipeline, param_grid, cv=3,
                           scoring='neg_root_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
cv_res = pd.DataFrame(grid_search.cv_results_)
cv_res.sort_values(by="mean_test_score", ascending=False, inplace=True)
cv_res.head()
```

# Get Final Model

```python
final_model = grid_search.best_estimator_  # includes preprocessing
feature_importances = final_model["random_forest"].feature_importances_
feature_importances.round(2)
```

```python
sorted(zip(feature_importances,
           final_model["preprocessing"].get_feature_names_out()),
           reverse=True)
```

```python
final_predictions = final_model.predict(X_test)

final_rmse = root_mean_squared_error(y_test, final_predictions)
print(final_rmse)
```
